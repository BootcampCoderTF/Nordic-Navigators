```python
# Import libraries
import pandas as pd
```

```python
# load first dataset to be modified from CSV
location_df = pd.read_csv('../Resources/Processed/locationsPriced.csv')
location_df
```

```python
# load second dataset to be modified from CSV
rating_df = pd.read_csv('../Resources/Processed/ratings.csv')
rating_df
```

```python
# Merge datasets on "Resort ID"
merged_df = pd.merge(location_df, rating_df, on="Resort ID")
merged_df
```

```python
# Convert merged DataFrame to JSON
json_data = merged_df.to_json(orient="records")
```

```python
# Save JSON data to file
with open("../DataVisualisationDropdowns/static/data/ratedResorts.json", "w") as json_file:
    json_file.write(json_data)
```
