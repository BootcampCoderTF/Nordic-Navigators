```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data from the CSV files
difficulty_path = '../Resources/Processed/difficulty.csv'
price_path = '../Resources/Processed/price.csv'

difficulty_data = pd.read_csv(difficulty_path)
price_data = pd.read_csv(price_path)
```

```python
merged_data = pd.merge(difficulty_data, price_data, on='Resort ID')
```

```python
# Sort the data by slope counts and select the top 10
top_10_difficulty = merged_data.nlargest(10, ['Easy', 'Intermediate', 'Difficult'])

# Create bar charts with increased size
plt.figure(figsize=(12, 10))
```

```python
# Beginner-friendly slope count compared to price
sns.barplot(x='Easy', y='Adult', data=top_10_difficulty, color='skyblue')
plt.title('Top 10 Beginner-friendly Slopes vs Price')
plt.xlabel('Height of Beginner-friendly Slopes (kms)')
plt.ylabel('Price')

plt.tight_layout()
plt.show()

# Save the plot as an image in the "Images" folder
plot_name = 'EasyVsCost'
plt.savefig(f'Images/{plot_name}.png')
plt.close()
```

```python
# Intermediate slope count compared to price
sns.barplot(x='Intermediate', y='Adult', data=top_10_difficulty, color='skyblue', ci=None)
plt.title('Top 10 Intermediate Slopes vs Price')
plt.xlabel('Height of Intermediate Slopes (kms)')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Save the plot as an image in the "Images" folder
plot_name = 'DifficultVsCost'
plt.savefig(f'Images/{plot_name}.png')
plt.close()
```

```python
# Difficult slope count compared to price
sns.barplot(x='Difficult', y='Adult', data=top_10_difficulty, color='skyblue', ci=None)
plt.title('Top 10 Difficult Slopes vs Price')
plt.xlabel('Height of Difficult Slopes (kms)')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Save the plot as an image in the "Images" folder
plot_name = 'DifficultVsCost'
plt.savefig(f'Images/{plot_name}.png')
plt.close()
```
